```python
import pandas as pd
import numpy as np
```

### Data Load

```python
df = pd.read_csv('/content/2. ecommerce sales data 2019.csv')
df.head()
```

```python
df.info()
```

### Data Cleaning & Prep

```python
df['Order Date'] = df['Order Date'].astype('str')
df.info()
```

```python
df['Order Date'].value_counts()
```

```python
df['Order Date'] = df['Order Date'].str.replace('^(\d{1,2}/\d{1,2}/)\d\d(\d\d)', r'\1\2',regex=True)
```

```python
df['Order Date']
```

```python
len_x = []
for i in df['Order Date']:
  len_xi = len(i)
  len_x.append(len_xi)
```

```python
len_x
```

```python
df['len_date'] = len_x
```

```python
df['len_date'].value_counts()
```

```python
df[df['len_date']==10]
```

```python
df = df[df['len_date']==14]
```

```python
df
```

```python
df['Quantity Ordered'] = df['Quantity Ordered'].astype('float')
df['Quantity Ordered'] = df['Quantity Ordered'].astype('int')
df['Price Each'] = df['Price Each'].astype('float')
```

```python
df.info()
```

```python
df['Revenue'] = df['Quantity Ordered']*df['Price Each']
```

```python
df.head()
```

### Sales performance over time, In what month do our sales peak or drop the most?


```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%y %H:%M')
```

```python
df.info()
```

```python
df.groupby(df['Order Date'].dt.month).agg({'Revenue': 'sum'}).reset_index(drop=False).rename(columns={'Order Date':'Month'})
```

### Which city has the top and lowest orders during 2019 


```python
df.head()
```

```python
df2019=df[df['Order Date'].dt.year==2019]
df2019
```

```python
df2019.groupby(df2019['City']).agg({'Quantity Ordered': 'sum'}).reset_index(drop=False).sort_values('Quantity Ordered', ascending=False)
```

### Which city contributed the most and the least to our revenue in 2019

```python
df2019=df[df['Order Date'].dt.year==2019]
df2019
```

```python
df2019.groupby(df2019['City']).agg({'Revenue': 'sum'}).reset_index(drop=False).sort_values('Revenue', ascending=False)
```
